```python
df["gdo_id"] = df["gdo_id"].astype("str")
```

```python
PARQUET_PATH = PATH + ".snappy.parquet"

df.to_parquet(PARQUET_PATH)
```

```python
def file_size(fname):
    import os
    statinfo = os.stat(fname)
    return statinfo.st_size

meg = 1024*1024
csv_size = file_size(PATH)
parquet_size = file_size(PARQUET_PATH)

print(f"CSV size: {int(csv_size/meg)}M")
print(f"Parquet size: {int(parquet_size/meg)}M {int(parquet_size/csv_size*100)}%")
```
