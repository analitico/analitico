```python
import pandas as pd
import numpy as np

ROWS = 15

def handle(event, context, **kwargs):
    # return a pandas dataframe directly, handler will return as json
    df = pd.DataFrame({
        "0Seq": list(range(0, ROWS)),
        "1Rand": np.random.randint(0, 100, ROWS),
        "2Nan": np.nan,
        "3None": None,
        "4Inf": np.inf
        })
    return df
```

```python
handle(None, None)
```
