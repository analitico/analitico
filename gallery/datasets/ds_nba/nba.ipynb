## NBA Player Salary Dataset (2017 - 2018)
An example of quick data profiling and analysis

```python
# to profile the dataset we use an external library which is not part
# of the default, conda based, platform so we need to install it first
# https://github.com/pandas-profiling/pandas-profiling#installation

!pip install pandas-profiling -q

import re
import pandas as pd
import pandas_profiling
```

```python
# read nba salaries from apache parquet dataset
df = pd.read_parquet("nba.parquet")

# would be similar if we had .csv data instead
# df = pd.read_csv("nba.csv")
```

```python
r = re.compile(r"([0-9]+)-([0-9]*\.?[0-9]+)")
def get_inches(el):
    m = r.match(el)
    if m == None:
        return float('NaN')
    else:
        return int(m.group(1))*12 + float(m.group(2))
      
df["Height"] = df["Height"].apply(lambda x: get_inches(x)) 
df["Height-metric"] = df["Height"] * 2.54
```

```python
df
```

```python
# profile dataset and output as html report
profile = df.profile_report(title='NBA Player Salary Dataset')
profile.to_file(output_file="nba-profiling.html")
```
