```
from datetime import datetime
def time_ms(started_on=None):
    """ Returns the time elapsed since given time in ms """
    return datetime.now() if started_on is None else int((datetime.now() - started_on).total_seconds() * 1000)
```

```
def csv_to_table(filename, tablename, engine):
  df = pd.read_csv(filename)
  df.set_index("id", inplace=True)
  df.to_sql(tablename, engine, if_exists="replace")
  return df

# if you upload a file in this notebook, you can create the table starting from the file
# csv_to_table("order.csv", "order", engine)


```

```
def table_to_csv(tablename, engine):
  started_on = time_ms()
  table = db.Table(tablename, metadata, autoload=True, autoload_with=engine)
  print("Reading " + tablename)
  print(table.columns.keys())
  df = pd.read_sql(table.select(), engine)
  df.set_index('id', inplace=True)
  filename = tablename + ".csv"
  df.to_csv(filename)
  print("Converted %d rows from %s to %s in %d ms" % (len(df), tablename, filename, time_ms(started_on)))
  return df, filename
```
